```python
import yfinance as yf
import datetime as dt
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

ma_1 = 30
ma_2 = 100

start = dt.datetime.now() - dt.timedelta(days=365 * 3)
end = dt.datetime.now()

data = yf.download('META', start=start, end=end)

data[f'SMA_{ma_1}'] = data['Adj Close'].rolling(window=ma_1).mean()
data[f'SMA_{ma_2}'] = data['Adj Close'].rolling(window=ma_2).mean()


data = data.iloc[ma_2:]

plt.plot(data['Adj Close'],label='Share Price', color='black')
plt.plot(data[f'SMA_{ma_1}'], label='ma1', color = 'orange')
plt.plot(data[f'SMA_{ma_2}'], label='ma2', color = 'red')
plt.legend(loc='upper left')
plt.show()
%matplotlib inline
```

```python
import yfinance as yf
import datetime as dt
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

buy_signals = []
sell_signals = []
trigger = 0


for x in range(len(data)):
  if data[f'SMA_{ma_1}'].iloc[x] > data[f'SMA_{ma_2}'].iloc[x] and trigger != 1:
    buy_signals.append(data['Adj Close'].iloc[x])
    sell_signals.append(float('nan'))
    trigger = 1

  elif data[f'SMA_{ma_1}'].iloc[x] < data[f'SMA_{ma_2}'].iloc[x] and trigger != 1:
    buy_signals.append(float('nan'))
    sell_signals.append(data['Adj Close'].iloc[x])
    trigger = -1

  else:
    buy_signals.append(float('nan'))
    sell_signals.append(float('nan'))


data['Buy'] = buy_signals
data['Sell'] = sell_signals
data.dropna(inplace=True)


plt.plot(data['Adj Close'],label='Share Price', color='black')
plt.plot(data[f'SMA_{ma_1}'], label='ma1', color = 'orange', linestyle = "--")
plt.plot(data[f'SMA_{ma_2}'], label='ma2', color = 'pink', linestyle = "--")

plt.scatter(data.index, data['Buy'], label='Buy', marker='^', color='green')
plt.scatter(data.index, data['Sell'], label='Sell', marker='v', color='red')
plt.legend(loc='upper left')
plt.show()
```

```python
import yfinance as yf
import datetime as dt
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

ma_1 = 30
ma_2 = 100

start = dt.datetime.now() - dt.timedelta(days=365 * 3)
end = dt.datetime.now()

# Update the ticker symbol from 'FB' to 'META'
df = yf.download('META', start=start, end=end)

# Calculate moving averages
df[f'SMA_{ma_1}'] = df['Adj Close'].rolling(window=ma_1).mean()
df[f'SMA_{ma_2}'] = df['Adj Close'].rolling(window=ma_2).mean()

# Initialize signals
buy_signals = np.where(df[f'SMA_{ma_1}'] > df[f'SMA_{ma_2}'], df['Adj Close'], np.nan)
sell_signals = np.where(df[f'SMA_{ma_1}'] < df[f'SMA_{ma_2}'], df['Adj Close'], np.nan)

# Plotting
plt.figure(figsize=(14,7))
plt.plot(df['Adj Close'], label='Adjusted Close Price', color='black')
plt.plot(df[f'SMA_{ma_1}'], label=f'SMA {ma_1}', color='orange', linestyle='--')
plt.plot(df[f'SMA_{ma_2}'], label=f'SMA {ma_2}', color='pink', linestyle='--')
plt.scatter(df.index, buy_signals, label='Buy Signal', marker='^', color='green')
plt.scatter(df.index, sell_signals, label='Sell Signal', marker='v', color='red')
plt.title('Buy and Sell Signals')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```
